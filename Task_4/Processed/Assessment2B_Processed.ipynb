###### Department wise codekata performence (pie chart)

```python
import matplotlib.pyplot as plt
dd=df.groupby(['Department'])['CodeKata Score'].sum().reset_index(name='sum')
plt.pie(dd['sum'],labels=dd['Department'],shadow=True)
plt.legend(title="Deparments")
plt.title("Department wise codekata performence")
plt.show()
```

###### Department wise toppers (horizantal bar graph or any visual representations of your choice)

```python
deps=df.groupby(['Department'])['CodeKata Score'].nlargest(10).reset_index(name='Toppers')
plt.barh(deps['Department'],deps['Toppers'])
plt.show()
```
